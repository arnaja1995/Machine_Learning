An Ecommerce company based in New York City, sells clothing online but they also have in-store style and clothing advice sessions. Customers come in to the store, have sessions/meetings with a personal stylist, then they can go home and order either on a mobile app or website for the clothes they want.

# Traning and Testing data

```python
X=customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
y=customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Training the Data

```python
from sklearn.linear_model import LinearRegression
LRM=LinearRegression()
```

```python
LRM.fit(X_train, y_train)
```

**Coefficients of the Model**

```python
print('Coefficients: \n', LRM.coef_)
```

**Coefficients DataFrame**

```python
coeff_df=pd.DataFrame(LRM.coef_,X.columns, columns=['Coefficients'])
coeff_df
```

**The Intercept**

```python
print('Intercept: \n', LRM.intercept_)
```

# Predicting Test Data

```python
predictions=LRM.predict(X_test)
```

```python
plt.scatter(y_test, predictions,marker='*',color='orange')
```

# Evaluating the Model

**Calculation of the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error**

```python
from sklearn import metrics
print('MAE: ',metrics.mean_absolute_error(y_test,predictions))
print('MSE: ',metrics.mean_squared_error(y_test,predictions))
print('RMSE: ',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

**Residuals**

```python
sns.set_palette('tab20c')
sns.displot((y_test-predictions),bins=50);
```

# Conclusion

**From the dataframe coeff_df we have the following:**

Interpretation of the coefficients:

- Holding all other features fixed, a 1 unit increase in **Avg. Session Length** is associated with an **increase of 25.98 total dollars spent**.
- Holding all other features fixed, a 1 unit increase in **Time on App** is associated with an **increase of 38.59 total dollars spent**.
- Holding all other features fixed, a 1 unit increase in **Time on Website** is associated with an **increase of 0.19 total dollars spent**.
- Holding all other features fixed, a 1 unit increase in **Length of Membership** is associated with an **increase of 61.27 total dollars spent**.

So the company need to develop the Website to catch up to the performance of the mobile app, or develop the app more since that is what is working better. In this case as the other factors play a crucial role, it is better to analyze the correlation between Length of Membership and App/Website performance.
